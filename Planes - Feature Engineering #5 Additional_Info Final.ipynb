```python
## libraries 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sqlite3
```

```python
# calling in the tables 

conn = sqlite3.connect(r"C:\Users\Epicm\Desktop\data_bases\dbs\planes\plane1.db")
cursor = conn.cursor()

# calling the flights table and saving as a df

query = '''

SELECT * from flights_cleaned

'''

df = pd.read_sql_query(query, conn)
conn.close()

df
```

```python
## no check-in baggage 

# this one just sounds weird in conversation
# I will be changing the values around to create check-in baggage instead 

df['additional_info check-in baggage included'] = np.where(df['additional_info_no check-in baggage included'] == 1, 0, 1)
df = df.drop('additional_info_no check-in baggage included', axis=1)

df
```

```python
# no info 

df = df.drop('additional_info_no info', axis=1)
df
```

```python
# changes airports 

condition = (df['total_stops'] >= 2) 
df.loc[condition, 'additional_info_change airports'] = 1

df
```
